## 04. Concistency Checks

### A. Mixed-Type Data

```python
# check for mixed-type columns
for col in dfhs.columns.tolist():
  weird = (dfhs[[col]].applymap(type) != dfhs[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (dfhs[weird]) > 0:
    print (col)
```

No mixed data types.
